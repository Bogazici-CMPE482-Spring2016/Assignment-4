###10.4. (b)###

First, we need to form a rotation matrix that will zero the lowermost left element. We form this matrix using the Givens rotation method, and call the matrix $G_1$. We can similarly form Givens matrices which will zero the other sub-diagonal elements in order, forming a triangular matrix R. The orthogonal matrix $Q^*$ is formed from the concatenation of all the Givens matrices $Q^* = G_n * G_{n-1} * ... * G_1$. Thus, we have $G_n * G_{n-1} * ... * G_1 * A = Q^* * A = R$, and the QR decomposition is $A = QR$.


```python
import random 
A = [[random.random() for j in range(3)] for i in range(3)]
G = np.identity(3) 

for i in range(len(A[0])):
    for j in range(len(A)-1,-1,-1):
        if i < j:
            theta = np.arctan(np.divide(-1*A[j][i],A[j-1][i]))
            if i == 0:
                G = np.dot([[np.cos(theta), -1*np.sin(theta), 0],[np.sin(theta), np.cos(theta), 0],[0,0,1]], G)
            elif i == 1:
                G = np.dot([[1,0,0],[0, np.cos(theta), -1*np.sin(theta)],[0, np.sin(theta), np.cos(theta)]], G)  
        


                        
                
            
print 'A\n\n',A
print '\nQ\n\n',np.transpose(G)
print '\nR\n\n',np.dot(G,A)
```
