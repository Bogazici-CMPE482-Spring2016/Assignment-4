```octave
h = ones(8,6); % H
h(1:3,3:4) = 0;
h(6:8,3:4) = 0;

e = ones(8,6); % E
e(3,3:6) = 0;
e(6,3:6) = 0;

l = ones(8,6); % L
l(1:6,3:6) = 0;

o = ones(8,6); % O
o(3:6,3:4) = 0;

A = zeros(15,40);
A(2:9,2:7) = h;
A(3:10,10:15) = e;
A(4:11,18:23) = l;
A(5:12,26:31) = l;
A(6:13,34:39) = o;

spy(A);
```

```octave
rnk = rank(A)
```

```octave
s = svd(A);
plot(s);
```

```octave
semilogy(s);
```

```octave
[U,S,V] = svd(A);
% A best k-rank approximation is given by zeroing out the r-k trailing singular values of A
SS = zeros(15,40);
```
