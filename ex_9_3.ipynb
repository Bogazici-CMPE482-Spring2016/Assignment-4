### a)

```matlab
A = zeros(15, 40);

% H
A(2:9, [2, 3, 6, 7]) = 1;
A(5:6, 4:5) = 1;

% E
A(3:10, 10:11) = 1;
A([3, 4, 6, 7, 9, 10], 12:15) = 1;

% L
A(4:11, 18:19) = 1;
A(10:11, 20:23) = 1;

% L
A(5:12, 26:27) = 1;
A(11:12, 28:31) = 1;

% O
A(6:13, 34:39) = 1;
A(8:11, 36:37) = 0;

spy(A);
```

### b)

```matlab
[U, S, V] = svd(A);

s = diag(S);
fprintf('Singular values:\n');
disp(s);
```

```matlab
figure('name', 'Singular Values Plot', 'numbertitle', 'off');
plot(s);
title('Singular Values Plot');
```

```matlab
figure('name', 'Singular Values Log-Scale Plot', 'numbertitle', 'off');
semilogy(s);
title('Singular Values Log-Scale Plot');
```

### Mathematically exact rank of A

```matlab
r = rank(A);
fprintf('rank(A): %d\n', r);
```

### Number of non-zero singular values of A is equal to the rank of A. Altough 11th and 12th singular values are not exactly zero, it is possible to observe from the plots (and as well as the values itself) that they are very close to zero thus can be discarded.

```matlab
fprintf('length(s(s>0)): %d\n', length(s(s>eps*2)));
```
