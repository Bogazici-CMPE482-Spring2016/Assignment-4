```matlab
m = 50;
n = 12;
t = linspace(0, 1, m);
A = fliplr(vander(t));
A = A(:, 1:12);
b = cos(4*t)';
```

```matlab
R_a = chol(A' * A);
w_a = R_a' \ (A' * b);
x_a = R_a \ w_a;
figure;
hold on;
plot(t, A * x_a, 'r', 'LineWidth', 2, 'MarkerSize', 5);
plot(t, b, 'b*', 'MarkerSize', 4, 'MarkerFaceColor', 'b');
```

```matlab
[Q_b, R_b] = mgs(A);
x_b = R_b \ (Q_b' * b);
figure;
hold on;
plot(t, A * x_b, 'r', 'LineWidth', 2, 'MarkerSize', 5);
plot(t, b, 'b*', 'MarkerSize', 4, 'MarkerFaceColor', 'b');
```

```matlab
[Q_c, R_c] = house(A);
x_c = R_c \ (Q_c' * b);
figure;
hold on;
plot(t, A * x_c, 'r', 'LineWidth', 2, 'MarkerSize', 5);
plot(t, b, 'b*', 'MarkerSize', 4, 'MarkerFaceColor', 'b');
```

```matlab
[Q_d, R_d] = qr(A);
x_d = R_d \ (Q_d' * b);
figure;
hold on;
plot(t, A * x_d, 'r', 'LineWidth', 2, 'MarkerSize', 5);
plot(t, b, 'b*', 'MarkerSize', 4, 'MarkerFaceColor', 'b');
```

```matlab
x_e = A \ b;
figure;
hold on;
plot(t, A * x_e, 'r', 'LineWidth', 2, 'MarkerSize', 5);
plot(t, b, 'b*', 'MarkerSize', 4, 'MarkerFaceColor', 'b');
```

```matlab
[U, S, V] = svd(A);
w_f = S \ (U' * b);
x_f = V' \ w_f;
figure;
hold on;
plot(t, A * x_f, 'r', 'LineWidth', 2, 'MarkerSize', 5);
plot(t, b, 'b*', 'MarkerSize', 4, 'MarkerFaceColor', 'b');
```

### We can see that all of the methods we tried yields with a nice approximation. If we dig deeper, we can see that there are some round-off errors by the nature of computer based computation. The normal equations are the fastest method whereas it produces the most inaccurate result as we can observe from the red fluctuations at the plot below.

```matlab
figure;
hold on;
plot(x_a, 'r');
plot(x_b, 'b');
plot(x_c, 'g');
plot(x_d, 'c');
plot(x_e, 'm');
plot(x_f, 'y');
```
